1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
a = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
mean = sum(a) / len(a)
var = sum(list(map(lambda x: (x - mean) ** 2, a))) / len(a)
var_b = sum(list(map(lambda x: (x - mean) ** 2, a))) / (len(a) - 1)
std = np.sqrt(var)
print(f'Среднее арифметическое: {mean} \nCреднее квадратичное отклонение:{std} '
      f'\nСмещенная дисперсия: {var} \nНесмещенная дисперсия: {var_b}')
```

2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

Вероятность того, что 3 мяча белые, складывается из следующих вероятностей:  

- Вероятность вытащить 2 белых мяча из первого ящика и 1 из второго  
- Вероятность вытащить 1 белый мяч из первого ящика и 2 из второго  
- Вероятность вытащить 0 белых мячей из первого ящика и 3 из второго  

$ \frac{C_5^2C_3^0}{C_8^2} \cdot \frac{C_5^1C_7^3}{C_{12}^4} + \frac{C_5^1C_3^1}{C_8^2} \cdot \frac{C_5^2C_7^2}{C_{12}^4} + \frac{C_5^0C_3^2}{C_8^2} \cdot \frac{C_5^3C_7^1}{C_{12}^4} = \frac{\frac{5!}{2!3!}}{\frac{8!}{2!6!}} \cdot \frac{\frac{5!}{4!}\cdot\frac{7!}{3!4!}}{\frac{12!}{4!8!}}+\frac{\frac{5!}{4!}\cdot\frac{3!}{2!}}{\frac{8!}{2!6!}} \cdot \frac{\frac{5!}{2!3!}\cdot\frac{7!}{2!5!}}{\frac{12!}{4!8!}}+\frac{\frac{3!}{2!}}{\frac{8!}{2!6!}} \cdot \frac{\frac{5!}{3!2!}\cdot\frac{7!}{6!}}{\frac{12!}{4!8!}} = \frac{5!2!6!5!7!4!8!}{2!3!8!4!3!4!12!}+\frac{5!3!2!6!5!7!4!8!}{4!2!8!2!3!2!5!12!}+\frac{3!5!7!2!6!4!8!}{2!3!2!6!8!12!} = \frac{5!5!6!7!}{3!3!4!12!}+\frac{5!6!7!}{2!2!12!}+\frac{4!5!7!}{2!12!} = \frac{25}{198}+\frac{5}{22}+\frac{1}{66} = \frac{73}{198} \approx 0.369 $

3. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится:  
a) на факультете A  
б) на факультете B  
в) на факультете C  

Студенты делятся по факультетам в следующей пропорции:  
$ A = \frac{1}{4} $, $ B = \frac{1}{4} $, $ C = \frac{2}{4} = \frac{1}{2} $  

Отсюда, вероятность  

$P_A = \frac{\frac{1}{4} \cdot 0.8}{\frac{1}{4} \cdot 0.8 + \frac{1}{4} \cdot 0.7 + \frac{1}{2} \cdot 0.9} = \frac{0.2}{0.2+0.175+0.45} \approx 0.24 $  

$P_B = \frac{\frac{1}{4} \cdot 0.7}{\frac{1}{4} \cdot 0.8 + \frac{1}{4} \cdot 0.7 + \frac{1}{2} \cdot 0.9} = \frac{0.175}{0.2+0.175+0.45} \approx 0.21 $  

$P_C = \frac{\frac{1}{2} \cdot 0.9}{\frac{1}{4} \cdot 0.8 + \frac{1}{4} \cdot 0.7 + \frac{1}{2} \cdot 0.9} = \frac{0.45}{0.2+0.175+0.45} \approx 0.55 $

4. Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя:  
а) все детали  
б) только две детали  
в) хотя бы одна деталь  
г) от одной до двух деталей   

$ p_1 = 0.1 $,  $ q_1 = 0.9 $  
$ p_2 = 0.2 $,  $ q_2 = 0.8 $  
$ p_3 = 0.25 $,  $ q_3 = 0.75 $  

a) $ P = p_1 \cdot p_2 \cdot p_3 = 0.1 \cdot 0.2 \cdot 0.25 = 0.005 $  

б) Только 2 детали состоят из вероятностей:  
- Выйдут из строя 1-я и 2-я, и не выйдет из строя 3-я  
- Выйдут из строя 1-я и 3-я, и не выйдет из строя 2-я
- Выйдут из строя 2-я и 3-я, и не выйдет из строя 1-я

$ P = p_1 \cdot p_2 \cdot q_3 + p_1 \cdot p_3 \cdot q_2 + p_2 \cdot p_3 \cdot q_1 = 0.1 \cdot 0.2 \cdot 0.75 + 0.1 \cdot 0.25 \cdot 0.8 + 0.2 \cdot 0.25 \cdot 0.9 = 0.08 $  

в) Хотя бы одна деталь может быть найдена через вероятность, что ни одна деталь не сломалась и вычитания этой вероятности из 1  

$ P = 1 - q_1 \cdot q_2 \cdot q_3 = 1-0.9 \cdot 0.8 \cdot 0.75 = 0.46 $  

г) 1 или 2 детали состоит из вероятности ровно 1 и ровно 2(ее мы нашли в пункте б)  

Только одна деталь состоит из вероятностей:  
- Выйдет из строя 1-я и не 2-я, и не  3-я 
- Выйдет из строя 2-я и не 1-я, и не  3-я 
- Выйдет из строя 3-я и не 1-я, и не  2-я 

$ P_1 = p_1 \cdot q_2 \cdot q_3 + p_2 \cdot q_1 \cdot q_3 + p_3 \cdot q_1 \cdot q_2 = 0.1 \cdot 0.8 \cdot 0.75 + 0.2 \cdot 0.9 \cdot 0.75 + 0.25 \cdot 0.9 \cdot 0.8 = 0.375 $  

Значит, вероятность выхода из строя 1й или 2х деталей:  

$ P = 0.375+0.08 = 0.455 $
